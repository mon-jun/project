```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/mon-jun/forhh/master/database.csv', sep=',')
df
```

```python
print("Latitude mean = " + str(df['Latitude'].mean()), end='\t')
print("Latitude median = " + str(df['Latitude'].median()), end='\t')
print("Latitude standard deviation = " + str(df['Latitude'].std()))

print("Longitude mean = " + str(df['Longitude'].mean()), end = '\t')
print("Longitude median = " + str(df['Longitude'].median()), end='\t')
print("Longitude standard deviation = " + str(df['Longitude'].std()))

print("Elevation mean = " + str(df['Elevation (Meters)'].mean()), end = '\t')
print("Elevation median = " + str(df['Elevation (Meters)'].median()), end='\t')
print("Elevation standard deviation = " + str(df['Elevation (Meters)'].std()))
```

The first question that poped in my interest was what dependence are volcanoes eruptions with latitude and longtitude. The top three latitudes are approximately -5(over 200 cases), 15(more than 220 cases), and 55(over 200 cases). As for longtitude, top three (125, 145, 155) would all have approximately same number of occasions(130+ each). Therefore, I suppose that by combining these latitudes and longtitudes we might get a more probable location where eruption might appear. In support to my words, I took a (55,155) coordinate, thus pointing in Kuril Islands, where erupting volcanoes are not rare events.

Speaking of countries, where volcanic eruptions, the biggest percentage(approximately 13,1%) of them took place in South America, then go Indonesea and Japan with Taiwan and Marianas sharing both 9.4 %, followed by Kamchatka and Mainland Asia(that has 8.8 %). In other regions significantly fewer cases were recorded in the passt 10,000 years.
